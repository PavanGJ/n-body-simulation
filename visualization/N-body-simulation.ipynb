```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
import os
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
```

```python
#Declare constants
dir_path = './data/velocity/' 
N = 2000
height = 10
width = 10
color = ['r','b','g']
```

```python
data = dict()
for i in range(N):
    file_path = os.path.abspath(os.path.join(dir_path, 'output_'+str(i)+'.csv'))
    local_data = pd.read_csv(file_path, header=None).as_matrix()[:,0:3]
    data[i] = pd.DataFrame({'x':local_data[:,0], 'y':local_data[:,1], 'z':local_data[:,2]})
```

```python
def update(num):
    update_data = data[num]
    graph._offsets3d = (update_data.x, update_data.y, update_data.z)
    title.set_text('N-body-simulation, time = %d'%num)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
title = ax.set_title('N-body-simulation')

fig.set_size_inches(height, width, True)

update_data = data[0]
graph = ax.scatter(update_data.x, update_data.y, update_data.z, marker = "o", depthshade=None, c=color)

anime = animation.FuncAnimation(fig, update, N - 1, interval = 50, blit = False)
HTML(anime.to_html5_video())
```

```python
writer = animation.FFMpegWriter(fps=48, bitrate = 2048)
anime.save('videos/sample.mp4', dpi = 200, writer = writer)
```
